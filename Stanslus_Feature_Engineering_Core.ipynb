```python
import pandas as pd
```

```python
# Load data
filename= '/content/drive/MyDrive/Data Science/Datasets/Machine Learning Data/bikeshare_train - bikeshare_train.csv'
df = pd.read_csv(filename)
# Show head
df.head()
```

```python
# Drop the two columns
df = df.drop(['casual', 'registered'], axis=1)
```

```python
df.head()
```

```python
#Transform date and save them as objects 
df['date'] = pd.to_datetime(df['datetime'])
```

```python
#Create new columns from the datae with new features and save them as objects
df['Month Name'] = df['date'].dt.month_name()
df['day of week'] = df['date'].dt.day_name()
df['Hour of day'] = df['date'].dt.hour
```

```python
df.head()
```

```python
#Drop 'datetime' and 'season'
df = df.drop(['datetime', 'season'], axis=1)
```

```python
#CELSIUS TO FAHRENHEIGHT
df['temp_Fahrenheit'] = df.apply(lambda x: (9/5)*x['temp']+32,axis=1)
```

```python
df.head()
```

```python
#Create a new column, 'temp_variance' that is the difference between 'temp' and 'atemp'.  Drop the 'atemp' column.
df['temp_variance'] = df['temp'] - df['atemp'] 
```
