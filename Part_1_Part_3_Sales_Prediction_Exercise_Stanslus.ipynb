```python
#import pandas
import pandas as pd
```

```python
filename = '/content/drive/MyDrive/Data Science/Datasets/sales_predictions.csv'
df_sales = pd.read_csv (filename)
df_sales.head()
```

```python
#Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#Histogram of COUNTS OF MAJOR ITEM TYPES SOLD
sns.histplot(data=df_sales, x="Item_Type", edgecolor='black',bins=100)
plt.xticks(rotation = 45)
plt.title('MAJOR ITEMS SOLD IN THE SUPERMARKET', fontsize =20)
plt.xlabel('Items Sold')
plt.ylabel('Frequency Counts ')
```

```python
#Boxplot of sales of major Oulets and Outlet types
sns.boxplot(x='Outlet_Size', y='Item_Outlet_Sales', data= df_sales)
plt.title('Average Sales by Outlet Size', fontsize =20)
plt.xlabel('Size of Outlet')
plt.ylabel('Mean sales')
```

```python
#Boxplot of sales of major Oulets and Outlet types
sns.boxplot(x='Outlet_Location_Type', y='Item_Outlet_Sales', data= df_sales)
plt.title('Average Sales by Outlet Location', fontsize =20)
plt.xlabel('Location of Outlet')
plt.ylabel('Mean sales')
```

```python
#Histogram of COUNTS OF MAJOR ITEM TYPES SOLD
sns.histplot(data=df_sales, x="Item_Fat_Content", y="Item_Outlet_Sales",bins=30)
plt.xticks(rotation = 45)
plt.title('Histogram of Sales by Fat Content', fontsize =20)
plt.xlabel('Fat content')
plt.ylabel('Frequency')
```

```python
#Establish how sales have been evolving over the years
plt.figure(figsize=(10,5))
sns.lineplot(x='Outlet_Establishment_Year', y='Item_Outlet_Sales', data=df_sales[:500])
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
plt.title('Mean ownership of cats and dogs', fontsize =20)
plt.xlabel('Cats or Dogs')
plt.ylabel('Means')
plt.grid()
```
